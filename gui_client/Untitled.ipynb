```python
from PIL import Image, ImageDraw
import numpy as np
from IPython.display import display

# Load the image
img = Image.open('images/pfp/dp_user.jpeg')

# Define the maximum allowed height
max_height = 100

# Check if the image height is greater than the maximum allowed height
if img.height > max_height:
    # Resize the image to the maximum allowed height while preserving aspect ratio
    w, h = img.size
    img = img.resize((int(w * max_height / h), max_height), Image.ANTIALIAS)

# Create the luminous image
w, h = img.size
lum_img = Image.new('L', [w, h], 0)
draw = ImageDraw.Draw(lum_img)
draw.pieslice([(0, 0), (w, h)], 0, 360, fill=255)
img_arr = np.array(img)
lum_img_arr = np.array(lum_img)
final_img_arr = np.dstack((img_arr, lum_img_arr))

# Display the final image
display(Image.fromarray(final_img_arr))
```

```python
from PIL import Image, ImageDraw
import numpy as np
from IPython.display import display

# Load the image
img = Image.open('images/pfp/dp_user.jpeg.jp')

# Define the maximum allowed height
max_height = 500

# Check if the image height is greater than the maximum allowed height
if img.height > max_height:
    # Resize the image to the maximum allowed height while preserving aspect ratio
    w, h = img.size
    img = img.resize((int(w * max_height / h), max_height), Image.ANTIALIAS)

# Create the luminous image
w, h = img.size
lum_img = Image.new('L', [w, h], 0)
draw = ImageDraw.Draw(lum_img)
draw.rectangle([(0, 0), (w, h)], fill=255)
img_arr = np.array(img)
lum_img_arr = np.array(lum_img)
final_img_arr = np.dstack((img_arr, lum_img_arr))

# Display the final image
display(Image.fromarray(final_img_arr))
```
